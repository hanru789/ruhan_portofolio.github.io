```python
from google_play_scraper import Sort, reviews
import pandas as pd
import numpy as np

app_id = 'id.flip'

result, _ = reviews(
    app_id,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=10000,
    filter_score_with=None
)

df = pd.DataFrame(result)

df.to_csv('flip_review.csv', index=False)
```
